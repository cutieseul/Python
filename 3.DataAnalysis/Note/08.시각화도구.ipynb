```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
# NaN을 전(Prev) 데이터로 채움 / 근처에 있는걸로 채움
# NaN 값을 전국으로 바꾸기
df = df.fillna(method='ffill')
df.head()
```

```python
df.info()
```

```python
df[df['전입지별'].isnull()]
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
# 전출지별이 서울이고 전입지별이 서울이 아닌 데이터

df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul.head(10)
```

```python
# 전출지별 칼럼 삭제

df_seoul.drop(['전출지별'],axis=1, inplace=True)
df_seoul

```

```python
# 전입지별 ->전입지로 컬럼 이름 변경
df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
len(df_seoul)
```

```python
# df_seoul 의 index를 전입지로 변경
df_seoul.set_index("전입지",inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one.index,sr_one.values)
plt.show()
```

```python
plt.plot(sr_one) # 데이터의 수가 적으므로 가능
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트제목 추가
plt.plot(sr_one)
plt.title("서울 -> 경기 인구 이동")

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

# 범례 표기
plt.legend(labels = ['서울->경기'], loc = 'best')
plt.show()
```

### Style 추가해보기

```python
# 그래프 꾸미기

# Style 서식 지정
plt.style.use('ggplot')

# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))


# 타이틀
plt.title("부산 -> 서울 인구 이동", size=20)

# x축 눈금 라벨 회전
plt.xticks(rotation='vertical')
# y축 눈금 라벨 회전하기
plt.yticks(rotation=45)


# 축 이름 추가
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

# Marker 표시
plt.plot(sr_one, marker="o", markersize=10)

# y축 범위 지정
plt.ylim(10000,80000) # 최소값. 최대값

# 화살표 표시
plt.annotate('',
              xy=(4.5,65000), # 화살표의 머리부분
              xytext=(2,22000), # 화살표의 꼬리부분
              xycoords='data', # 좌표 체계
              arrowprops=dict(arrowstyle='->', color='skyblue', lw=5)

)

plt.annotate('',
              xy=(48,25000), # 화살표의 머리부분
              xytext=(13,60000), # 화살표의 꼬리부분
              xycoords='data', # 좌표 체계
              arrowprops=dict(arrowstyle='->', color='green', lw=5)

)

plt.annotate('인구 이동 증가(1972~1975)', # 텍스트
             xy = (2.5,35000), # 텍스트 위치 기준점
             rotation=78, # 텍스트 회전 각도
             va='baseline', # 텍스트 상하 정렬
             ha='center', # 텍스트 좌우 정렬
             fontsize=9 # 텍스트크키
)
plt.annotate('인구 이동 감소(1983~2017)', # 텍스트
             xy = (30,45000), # 텍스트 위치 기준점
             rotation=-13, # 텍스트 회전 각도
             va='baseline', # 텍스트 상하 정렬
             ha='center', # 텍스트 좌우 정렬
             fontsize=9 # 텍스트크키
)

# 범례 표기
plt.legend(labels = ['부산->서울'], loc = 'best', fontsize = 15)

# 이미지 저장
plt.savefig('../Data/seoul.png')

plt.show()
```

### 동일한 그림에 여러개의 그래프를 추가

```python
# 서울에서 충청남도, 경상북도, 강원도 이동한 데이터만 추출하여 정리
df_3 = df_seoul.loc[['충청남도','경상북도','강원도']]
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(df_3.columns,
        df_3.loc['충청남도'],
        marker='o',
        markerfacecolor='green',
        markersize=10,
        color='olive',
        linewidth=2,
        label='서울 -> 충남')

ax.plot(df_3.columns,
        df_3.loc['경상북도'],
        marker='o',
        markerfacecolor='blue',
        markersize=10,
        color='skyblue',
        linewidth=2,
        label='서울 -> 경북')

ax.plot(df_3.columns,
        df_3.loc['강원도'],
        marker='o',
        markerfacecolor='red',
        markersize=10,
        color='magenta',
        linewidth=2,
        label='서울 -> 강원')

ax.legend(loc='best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)

ax.set_xlabel('기간',size=12)
ax.set_ylabel('이동 인구수',size=12)

ax.set_xticklabels(df_3.columns, rotation=90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x',labelsize=9)
ax.tick_params(axis='y',labelsize=9)


plt.show()
```

```python
# 서울에서 충청남도, 경상북도, 강원도 이동한 데이터만 추출하여 정리
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4.columns
```

```python
# 그래프 꾸미기

# 스타일 서식 지정
plt.style.use("ggplot")

# 그래프 객체 생성(figure에 2개의 서브플롯을 생성)
fig = plt.figure(figsize=(20,12)) # 그래프 사이즈
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(
        df_4.loc['충청남도'],
        marker='o',
        markerfacecolor='green',
        markersize=10,
        color='olive',
        linewidth=2,
        label='서울 -> 충남') 

ax2.plot(
        df_4.loc['경상북도'],
        marker='o',
        markerfacecolor='blue',
        markersize=10,
        color='skyblue',
        linewidth=2,
        label='서울 -> 경북') 

ax3.plot(
        df_4.loc['강원도'],
        marker='o',
        markerfacecolor='red',
        markersize=10,
        color='magenta',
        linewidth=2,
        label='서울 -> 깅원') 

ax4.plot(
        df_4.loc['전라남도'],
        marker='o',
        markerfacecolor='orange',
        markersize=10,
        color='yellow',
        linewidth=2,
        label='서울 -> 전남') 




ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_title('서울 -> 충남 인구이동')
ax2.set_title('서울 -> 경북 인구이동')
ax3.set_title('서울 -> 강원 인구이동')
ax4.set_title('서울 -> 전라 인구이동')

ax1.set_ylim(0,70000)
ax2.set_ylim(0,70000)
ax3.set_ylim(0,70000)
ax4.set_ylim(0,70000)

ax1.set_xticklabels(df_4.columns, rotation=75)
ax2.set_xticklabels(df_4.columns, rotation=75)
ax3.set_xticklabels(df_4.columns, rotation=75)
ax4.set_xticklabels(df_4.columns, rotation=75)

plt.show()
```
