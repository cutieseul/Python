```python
import nltk
import pickle
from nltk.corpus import stopwords
import re
nltk.download('all')
```

### text를 불러서 word분류

```python
content = open('../Data/Independence.txt').read()
content
```

```python
# 불필요한 심벌 없애기
# cleaned_content = content.replace('!', '').replace(",",'')
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]','',content) # 문장 단위로 끊기
cleaned_content
```

```python
# 대문자를 소문자로 변환
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 문장에서 위드를 토큰으로 쪼갠다.
word_tokens =nltk.word_tokenize(cleaned_content)
print(word_tokens)
```

```python
# 품사분리
token_pos = nltk.pos_tag(word_tokens)
print(token_pos)
```

```python
# 명사만 추출하기
NN_words = []
for word, pos in token_pos:
    if 'NN' in pos:
        NN_words.append(word)

print(NN_words)
```

```python
# 원형으로 복원
wlem = nltk.WordNetLemmatizer()
lenmatized_works = []
for word in NN_words:
    new_word = wlem.lemmatize(word)
    lenmatized_works.append(new_word)

print(lenmatized_works)
```

```python
# 불용어 제거
stopwords_list = stopwords.words('english')
unique_NN_words = set(lenmatized_works)
final_NN_words = lenmatized_works
# unique는 1차원에서만 쓰기 때문에 set을 써주는 것. ! 다시한번 확인하기

# 중복값을 제거하려고 
for word in unique_NN_words:
    if word in stopwords_list:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 원하는 단어 제거
customized_stopwords = ['be', 'today', 'yesterday', "it's", "don't",'others']
unique_NN_words1 = set(final_NN_words)
for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 빈도 분석
from collections import Counter

c=Counter(final_NN_words)
print(c)
```

```python
print(c.most_common(20))
```

```python
# 워드 클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud

noun_text = ''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word

noun_text
```

```python
wordcloud = WordCloud(
    max_font_size=60,
    relative_scaling=0.5
).generate(noun_text) # generate가 noun_text의 string value를 입력받아 count함. 

plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
