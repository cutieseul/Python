```python
# index이름을 정하기
exam_data = {'수학':[90,80,70], '영어':[98,89,95],
 '음악':[85,95,100], '체육':[100,90,90], '이름': ['서준','우현','인아'] }
df = pd.DataFrame(exam_data)
df
```

```python
df.set_index('이름',inplace=True)
df
```

```python
# 서준의 음악 체육만 보기
print(df.loc['서준',['음악','체육']])
print()

# print(df.iloc[0,[2,3]]) 아래꺼랑 같음
print(df.iloc[0,2:4])

print()
print(df.loc['서준','음악':])
```

```python
df['국어'] = 80
df
```

```python
df.loc['철수'] = [10,20,30,40,50]
df
```

```python
# 영희 인덱스 새로 만들고 인아랑 같은 점수 넣기
df.loc['영희'] = df.loc['인아']
df
```

```python
# 서준의 체육점수를 80점으로 변경하기
df.loc['서준','체육'] = 80
df
```

```python
# 서준의 음악과 체육점수를 50점으로 변경하기
df.loc['서준',['체육','음악']] = 50
df
```

```python
# 서준의 음악 점수를 100점 체육점수를 50점으로 변경하기
df.loc['서준',['체육','음악']] = [50,100]
df
```

```python
# 행과 열 바꾸기
df.transpose()
# df.T 랑 똑같음
```

# DataFrame 연산

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 칼럼 유추 : 특성 공학 : column engineering
```

```python
# titanic에서 age 와 fare 부분의 칼럼만 발췌하여 df 라는 dataFrame 만들기
df = pd.DataFrame([titanic.age,titanic.fare]).T
df.head()
```

```python
df = titanic[['age','fare']]
df.head()
```

```python
df = titanic.loc[:,['age','fare']]
df.head()
```

```python
# df.describe()

# 모든 칼럼을 10 씩 증가하여 addition이라는 DataFrame 생성하기
addition = df + 10
addition
```

```python
# addition과 df의 차를 구하여 subtraction이라는 DataFrame 만들기
subtraction = addition - df
subtraction
```
