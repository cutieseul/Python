### CCTV 현황

```python
# CCTV에서 소계로 오름차순 정렬
CCTV_Seoul.sort_values(by='소계', inplace=True)
CCTV_Seoul.head()
```

> CCTV 댓수가 적은 지역으로는 도봉구, 중랑구, 강북구, 종로구, 중구 순이다.

```python
# CCTV에서 소계로 내림차순 정렬
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

> CCTV댓수가 많은 지역으로는 강남구, 관악구, 구로구, 성북구, 은평구 순이다

```python
#### 최근 3년간 CCTV증가율 구해서 CCTV 제일 많이 증가한 구를 검색후 정렬
CCTV_Seoul['최근증가율'] = CCTV_Seoul.iloc[:,6:].sum(axis=1) / CCTV_Seoul.iloc[:,2:7].sum(axis=1) *100
```

```python
CCTV_Seoul.sort_values('최근증가율', ascending=False).head()
```

> 결과를 보면 최근 3년간의 CCTV가 2015년 이전에 비해 가장 많이 증가한 곳은 중구, 영등포구, 성동구, 금천구, 광진구 순이다.

### 인구현황

```python
# 구별 Data의 유효성 확인
len(pop['구별'].unique())
```

```python
pop.head()
```

```python
pop['외국인비율'] = pop['외국인'] / pop['인구수'] *100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] *100
```

```python
pop.head()
```

```python
# 인구수가 제일 많은 지역은 Top5
pop.sort_values(by='인구수', ascending=False).head()
```

```python
# 외국인이 제일 많은 지역은 Top5
pop.sort_values(by='외국인', ascending=False).head()
```

```python
# 외국인 비율이 높은 지역은 Top5
pop.sort_values(by='외국인비율', ascending=False).head()
```

```python
# 고령자 정렬
pop.sort_values(by='고령자', ascending=False).head()
```

```python
# 고령자 비율 정렬
pop.sort_values(by='고령자비율', ascending=False).head()
```

3조
> 영등포구가 외국인인구수, 외국인 비율이 가장 높게 나타나는 경향을 보인다.      
> 영등포구, 금천구, 구로구는 외국인 인구수와 외국인비율이 높게 나타나는 경향을 보인다.
> 은평구를 제외한 나머지 구는 고령자인구와 고령자비율이 비례하지 않는다.      
> 인구수가 많은 상위 다섯개 구 중에 4개의 구가 고령자수가 높게 나타나는 경향을 보인다.(즉 인구수와 고령자수는 비례관계를 보인다.)            
> 고령자 비율은 강북지역이 높다.      
> 인구수 상위 5개 지역구 중 4군데가 강남지역이다.


인구수가 제일 많은 지역은 송파구이지만 영등포구가 제일 외국인이 많다.   
그리고 그 근처라고 할 수 있는 구로구, 금천구 등도 굉장히 외국인이 많다.       
송파구는 고령자가 제일 많지만 고령자의 비율이 높지 않은 것이 확인된다.    
하지만 강북구, 도봉구, 중구, 종로구, 은평구 같은 경우는       
인구수 대비 고령자 비율이 많기 때문에 추후 고령화가 우려된다.

> 역상관으로 보인다.

```python
# 외국인 비율과 소계의 상관계수
np.corrcoef(data_reault['외국인비율'],data_reault['소계'])
```

> 상관이 전혀 없어보인다.

```python
# 외국인과 소계의 상관계수
np.corrcoef(data_reault['외국인'],data_reault['소계'])
```

> 상관이 없어보인다.

```python
# 인구수와 소계의 상관계수
np.corrcoef(data_reault['인구수'],data_reault['소계'])
```

> 고령자 비율의 상관계수를 보면 역상관이 발생하고 이는 고령자가 아닌 젊은층 비율이 높은 곳에 CCTV가 많이 설치되는 경향을 보이며
> 주로 인구수와 양의 상관을 보이고 있어 CCTV 댓수는 인구수에 비례하여 증가하는 경향을 보인다.
