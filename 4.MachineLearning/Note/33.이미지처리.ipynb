# Numpy를 Image로 변환하기


```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Numpy Data  불러오기
import numpy as np
fruits = np.load("../Data/fruits.npy")
fruits.shape
```

```python
# Image로 저장하기
from PIL import Image
for i in range(0,300):
    image = Image.fromarray(fruits[i]).convert("L")
    image.save("../Data/fruits/fruit_%03d.jpg"%i,"JPEG")
```

# Image를 읽어서 Numpy로 변환하기

```python
import glob
```

```python
numnerOfData = 300
imgSize = 100
```

```python
# 배열 초기화로 만들기  (300 * 100 * 100)
fruitsData = np.zeros(numnerOfData * imgSize * imgSize, dtype=np.int32).reshape(numnerOfData, imgSize, imgSize)
```

```python
i = 0
for file in sorted(glob.glob("../Data/fruits/*.jpg")):
    img = np.array(Image.open(file), dtype=np.int32)
    fruitsData[i,:,:] = img
    i += 1
```
