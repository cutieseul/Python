```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://movie.daum.net/ranking/boxoffice/weekly"
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
li_lists = soup.select("#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a")
```

```python
li_lists = soup.select(".link_txt")
li_lists
```

```python
li_lists
```

```python
# 데이터 프레임을 만들려면 일단 list 로 만들자
    # R 처럼 list를 만들고 데이터프레임을 만들기

count = 0
titleLists = [] # 영화 제목 List
countLists = [] # 순위 List

for list in li_lists:
    count += 1
    print(count, list.string.strip())
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
# List 확인

print(countLists)
print(titleLists)
```

```python
# countLists만으로 DataFrame 생성

df = pd.DataFrame(data=countLists, columns=['순위'])
df
```

```python
# 기존 DataFrame에 컬럼과 List 추가
    # 파이썬에선 벡터값을 Series라고 한다.
df['영화제목'] = pd.Series(titleLists)
df
```

```python
# CSV로 저장하기
    # 중요! 인덱스는 안쓸거니까 False로 한다
df.to_csv("../Data/DaumList.csv", encoding='utf-8', index= False)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
naver_lists = soup.select("#old_content > table > tbody > tr > td.title > div > a")
naver_lists
```

```python
# 데이터 프레임을 만들려면 일단 list 로 만들자
    # R 처럼 list를 만들고 데이터프레임을 만들기

count = 0
titleLists = [] # 영화 제목 List
countLists = [] # 순위 List

for list in naver_lists:
    count += 1
    print(count, list.string.strip())
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
# List 확인

print(countLists)
print(titleLists)
```

```python
# countLists만으로 DataFrame 생성

df = pd.DataFrame(data=countLists, columns=['Index'])
df
```

```python
# 기존 DataFrame에 컬럼과 List 추가
    # 파이썬에선 벡터값을 Series라고 한다.
df['Titles'] = pd.Series(titleLists)
df
```

```python
# CSV로 저장하기
    # 중요! 인덱스는 안쓸거니까 False로 한다
df.to_csv("../Data/naverRanking.csv", encoding='utf-8', index= False)
```
