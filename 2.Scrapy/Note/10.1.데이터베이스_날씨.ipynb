```python
# Library
import requests
from bs4 import BeautifulSoup
import urllib.request as req

# Site Address
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 도시 출력
for i in soup.find_all("location"):
  print(i.find("city"))
```

```python
# 서울로 테스트
seoul = soup.find("location")
for i in seoul.find_all("data"):
  print(i.find("tmef"))
  print(i.find("wf"))
  print(i.find("tmn"))
  print(i.find("tmx"))
```

```python
# 도시별 날씨 데이터 구축
weather = {}

for i in soup.find_all("location"):
  # 키값을 만듦
  weather[i.find("city").text] = []
  for j in i.find_all("data"):
    temp = []
    temp.append(j.find("tmef").text)
    temp.append(j.find("wf").text)
    temp.append(j.find("tmn").text)
    temp.append(j.find("tmx").text)
    weather[i.find("city").text].append(temp)
```

```python
# 키값 입력 하면 그 값만 나옴
print(weather['서울'])
```

```python
# 현재 Dictionary 로 되어 있어 List 형태로 변환이 필요
tempList = []
for i in weather:
  for j in weather[i]:
    tempList.append([i,j[0],j[1],j[2],j[3]])
```

```python
import pandas as pd
```

```python
weatherDataFrame = pd.DataFrame(tempList, columns=['city','tmef','wf','tmn','tmx'])
weatherDataFrame.head()
```

```python
# csv로 저장
weatherDataFrame.to_csv("../Data/weather.csv", index=False)
pd.read_csv("../Data/weather.csv").head()
```
